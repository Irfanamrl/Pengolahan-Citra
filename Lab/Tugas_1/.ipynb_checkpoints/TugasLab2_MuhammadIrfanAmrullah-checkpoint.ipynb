```python
from skimage import color, io
from skimage.transform import rescale
import matplotlib.pyplot as plt
import numpy as np
```

```python
i1 = io.imread('ueno.jpg')
```

```python
plt.imshow(i1, vmin=0, vmax=255)
plt.show()
```

```python
e1 = color.rgb2gray(i1)
plt.imshow(e1, cmap = 'gray')
plt.show()
```

```python
from skimage import util
gray = util.img_as_ubyte(color.rgb2gray(i1))
plt.subplot(1,2,1); plt.hist(gray.flatten(), 256, range=(0,256))
plt.title('Original');
plt.show()
#show the histogram of image i1
```

```python
from skimage import exposure, util
gray = util.img_as_ubyte(e1)
plt.subplot(); plt.hist(gray.flatten(), 256, range=(0,256))
plt.title('Original');
plt.show();
gray1 = exposure.equalize_hist(e1)
gray_he = util.img_as_ubyte(gray1)
plt.subplot(); plt.hist(gray_he.flatten(), 256, range=(0,256))
plt.title('Histogram Equalization');
plt.show();
plt.imshow(gray, cmap='gray')
plt.title('Original'); plt.axis('off')
plt.show();
plt.imshow(gray_he, cmap='gray')
plt.title('Histogram Equalization'); plt.axis('off')
plt.show();
```

```python
from skimage import util

mn = min(e1.flatten())
mx = max(e1.flatten())
b = int(np.floor(255 / (mx - mn)))
e1_cs = (e1 - mn) * b

plt.subplot(); plt.imshow(e1, cmap='gray')
plt.title('Original'); plt.axis("off")
plt.show();
plt.subplot(); plt.imshow(e1_cs, cmap='gray')
plt.title('Contrast Stretching'); plt.axis("off")
plt.show();

gray = util.img_as_ubyte(e1)
plt.subplot(); plt.hist(gray.flatten(),256,range=(0,256))
plt.title('Original');
plt.show();
gray2 = util.img_as_float(e1_cs)
plt.subplot(); plt.hist(gray2.flatten(),256,range=(0,256))
plt.title('Contrast Stretching');
plt.show();

```

```python
from skimage import filters, morphology
gray3 = eq_i3
fi = filters.rank.mean(gray3, selem=morphology.square(2))
sh = filters.unsharp_mask(fi,radius=(2))
plt.subplot();plt.imshow(fi,cmap='gray',vmin=0,vmax=255)
```

```python
i4 = io.imread('teles2010.jpg')
i5 = io.imread('teles2000.jpg')
```

```python
i6 = np.subtract(i4, i5, dtype='int16')
e2 = color.rgb2gray(i6)
plt.subplot(); plt.imshow(e2, cmap='gray');
plt.title('After Substraction 2010 with 2000')
plt.show()
```

```python
a = e2/255
c1 = 3
f1 = c1*np.log(1 + (a))
plt.subplot(); plt.imshow(f1, cmap='gray')
plt.title('Log Transformation'); plt.axis("off")
plt.show();
```

```python
mn = min(f1.flatten())
mx = max(f1.flatten())
b = int(np.floor(255 / (mx - mn)))
f1_cs = (f1 - mn) * b

plt.subplot(); plt.imshow(f1_cs, cmap='gray')
plt.title('Contrast Stretching'); plt.axis("off")
plt.show();
```

```python
from skimage.feature import canny
edges = canny(f1_cs, sigma=5)
plt.subplot(); plt.imshow(f1_cs, cmap='gray')
plt.title('Canny'); plt.axis("off")
plt.show();
```

```python
from skimage.filters import threshold_otsu, threshold_minimum
thresh = threshold_minimum(f1_cs)
binary = f1_cs > thresh

plt.subplot(); plt.imshow(binary, cmap='gray')
plt.title("After Threshold"); plt.axis("off")
plt.show();
```

```python
i6 = io.imread('pantai1.jpg')
i7 = io.imread('pantai2.jpg')

i8 = np.subtract(i6, i7, dtype='int16')
e3 = color.rgb2gray(i8)
plt.subplot(); plt.imshow(e3, cmap='gray');
plt.title('After Substraction Pantai1 with Pantai2'); plt.axis("off")
plt.show()
```

```python
from skimage.filters import threshold_otsu, threshold_yen
thresh = threshold_otsu(e3)
binary = e3 < thresh

plt.subplot(); plt.imshow(binary, cmap='gray')
plt.title("After Threshold"); plt.axis("off")
plt.show();
```
